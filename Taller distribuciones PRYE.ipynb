## Analizar la convergencia de la ley de los grandes numeros

# Variable aleatoria normal

```python
import numpy as np
import matplotlib.pyplot as plt

n_muestras = 1000
pasos = range(1, n_muestras + 1)

# Combinaciones de media y desviacin estndar
parametros = [
    (0, 1),
    (5, 1),
    (-3, 1),
    (0, 2),
    (0, 5),
    (5, 2)
]

plt.figure(figsize=(14, 10))
for i, (mu, sigma) in enumerate(parametros, 1):
    medias_muestrales = []
    for n in pasos:
        muestra = np.random.normal(mu, sigma, n)
        media_muestral = np.mean(muestra)
        medias_muestrales.append(media_muestral)

    plt.subplot(3, 2, i)
    plt.plot(pasos, medias_muestrales, label=f"Media muestral")
    plt.axhline(mu, color='red', linestyle='--', label=f"Media real = {mu}")
    plt.title(f"mu = {mu}, sigma = {sigma}")
    plt.xlabel("Tamao de muestra")
    plt.ylabel("Media")
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.suptitle("Convergencia de la media muestral para distintas normales", fontsize=16, y=1.02)
plt.show()

```

### Conclusiones

La media muestral siempre tiende a la media real () cuando el tamao de muestra crece.

Una mayor desviacin estndar () provoca ms variacin en muestras pequeas.

Variar la muestra, la media o la desviacin, no afecta la normalidad de la distribucin.

La Ley de los Grandes Nmeros se cumple sin importar los parmetros, siempre que la distribucin sea normal y tenga media y varianza finitas.

# Variable aleatoria binomial

```python
import numpy as np
import matplotlib.pyplot as plt

n_muestras = 1000
pasos = range(1, n_muestras + 1)

# Combinaciones (n, p) para la distribucin binomial
parametros = [
    (10, 0.3),
    (10, 0.5),
    (10, 0.8),
    (50, 0.5),
    (100, 0.1),
    (100, 0.9)
]

plt.figure(figsize=(14, 10))
for i, (n, p) in enumerate(parametros, 1):
    medias_muestrales = []
    media_teorica = n * p
    for t in pasos:
        muestra = np.random.binomial(n, p, t)
        media_muestral = np.mean(muestra)
        medias_muestrales.append(media_muestral)

    plt.subplot(3, 2, i)
    plt.plot(pasos, medias_muestrales, label="Media muestral")
    plt.axhline(media_teorica, color='red', linestyle='--', label=f"Media real = {media_teorica}")
    plt.title(f"Binomial(n={n}, p={p})")
    plt.xlabel("Tamao de muestra")
    plt.ylabel("Media")
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.suptitle("Convergencia de la media muestral (Distribucin Binomial)", fontsize=16, y=1.02)
plt.show()
```

## Conclusiones

La media muestral converge a la media terica de la binomial, que es  = n*p, cuando la muestra tiende a infinito.

La variabilidad de la media muestral es mayor al inicio, especialmente si p est cerca de 0 o 1, pero disminuye con ms muestras.

La varianza de la binomial ( = n* p*(1-p)) influye en qu tan dispersas son las medias muestrales al principio. 

Sin importar los valores de n y p, la ley de los grandes numeros se cumple, siempre y cuando la varianza no sea infinita.

Aunque la distribucin binomial es discreta, la media muestral se comporta como una variable continua al aumentar el tamao de muestra.
